```python
import pandas as pd
from underthesea import word_tokenize, ner

# Step 1: 엑셀 파일에서 데이터 읽어오기
excel_file = 'youtube_vietnam_15_video.xlsx'  # 사용할 엑셀 파일 경로나 이름에 맞게 수정
df = pd.read_excel(excel_file)

# Step 2: 'comment' 열의 NaN 값 처리하기
df['comment'] = df['comment'].fillna('').astype(str)  # NaN 값을 빈 문자열로 채우고 문자열로 변환

# Step 3: Underthesea를 사용하여 단어 토큰화 및 개체명 인식(NER)을 준비하는 함수 정의
def tokenize_and_ner(text):
    words = word_tokenize(text)  # 단어 토큰화
    entities = ner(text, deep=True)  # 개체명 인식
    return words, entities

# Step 4: 각 댓글을 토큰화하고 명사를 추출하여 빈도수 계산
tokenized_comments = []
for comment in df['comment']:
    words, entities = tokenize_and_ner(comment)
    tokenized_comments.extend(words)

# Step 5: 단어의 빈도수 계산
word_freq = {}
for word in tokenized_comments:
    if word in word_freq:
        word_freq[word] += 1
    else:
        word_freq[word] = 1

# Step 6: 빈도수 기준으로 단어 정렬
sorted_word_freq = sorted(word_freq.items(), key=lambda x: x[1], reverse=True)

# Step 7: 엑셀 내보내기용 데이터 준비
rank = list(range(1, len(sorted_word_freq) + 1))
word_data = {'Rank': rank, 'Word': [word[0] for word in sorted_word_freq], 'Frequency': [word[1] for word in sorted_word_freq]}
df_words = pd.DataFrame(word_data)

# Step 8: 데이터프레임을 엑셀 파일로 저장
output_file = 'youtube_vietnam_word_frequencies.xlsx'
df_words.to_excel(output_file, index=False)

print(f"단어 빈도수가 {output_file}에 저장되었습니다.")
```
